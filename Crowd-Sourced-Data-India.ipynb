```python
r=pd.read_csv('https://api.covid19india.org/csv/latest/case_time_series.csv')
```

```python
plt.plot(range(0,len(r['Daily Confirmed'])),r['Daily Confirmed'],label='New Cases')

r['Removed'] = r['Daily Recovered'] + r['Daily Deceased']
plt.plot(range(0,len(r['Removed'])),r['Removed'],label='Deaths + Recovery (Removed)')
plt.xlabel('Days Since R0')
plt.ylabel('Cases (Linear)')
plt.yscale('log')
plt.title('Daily Increase (India)')
legend = plt.legend()
plt.savefig('jaihind2.png')
```

```python
r['Removed']
```

```python
r['Daily Active Confirms'] = r['Daily Confirmed'] - (r['Daily Recovered'] + r['Daily Deceased'])
r['Daily Recovered'] = (r['Daily Recovered'] + r['Daily Deceased'])
r['Daily Removed'] = r['Daily Recovered'] + r['Daily Deceased']
```

```python
plt.title('India  - New Cases')
plt.plot(range(0,len(r['Daily Active Confirms'])),r['Daily Active Confirms'],label='Daily Active Confirms')
plt.plot(range(0,len(r['Daily Recovered'])),r['Daily Recovered'],label='Daily Recovery')
#plt.yscale('log')
plt.xlabel('Days since P0')
plt.ylabel('Cases (Uniform Scale)')
legend = plt.legend()
```

```python
r['Daily Active Confirms']==r['Daily Removed']
```

```python
r['active'] = r['Total Confirmed'] - (r['Total Recovered'] + r['Total Deceased'])
```

```python
plt.plot(range(0,len(r['active'])),r['active'],label='Total Active Cases')
plt.plot(range(0,len(r['Total Recovered'])),r['Total Recovered'],label='Total Removal (Recovery + Deaths)')
legend = plt.legend()
plt.title('India - New Case Rate')
plt.xlabel('Days since P0')
plt.ylabel('Cases (Uniform Scale)')
plt.yscale('log')
```
