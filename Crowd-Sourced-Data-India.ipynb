```python
b= pd.read_csv('https://api.covid19india.org/csv/latest/case_time_series.csv')
```

```python
b
```

```python
b['Date'] = pd.to_datetime(b['Date'])
```

```python
plt.plot(range(0,len(b['Total Confirmed'])),b['Total Confirmed'],label='Total Confirmed')
plt.plot(range(0,len(b['Total Recovered'])),b['Total Recovered'],label='Total Recovered')
plt.plot(range(0,len(b['Total Deceased'])),b['Total Deceased'],label='Total Deceased')
plt.yscale('log')
plt.xlabel('Days Since R0')
plt.ylabel('Cases (Log Scale)')
legend = plt.legend()
plt.title('Republic of India')
plt.savefig('jaihind.png')
```

```python
r=pd.read_csv('https://api.covid19india.org/csv/latest/case_time_series.csv')
```

```python
plt.plot(range(0,len(r['Daily Confirmed'])),r['Daily Confirmed'],label='New Cases')

r['Removed'] = r['Daily Recovered'] + r['Daily Deceased']
plt.plot(range(0,len(r['Removed'])),r['Removed'],label='Deaths + Recovery (Removed)')
plt.xlabel('Days Since R0')
plt.ylabel('Cases (Linear)')
plt.title('Daily Increase (India)')
legend = plt.legend()
plt.savefig('jaihind2.png')
```
